# Waves

## Introduction

+ Before we continue any further, we have to get to grips with waves
+ A large part of quantum mechanics revolves around the wave equation of Schr&ouml;dinger

## The unit circle

+ In physics, it is much easier to work with radians than with degrees
+ Radians is an *imaginary* unit, i.e. we write it down, but it is not a true dimension; it is *dimensionless*
+ We know that *&pi;* is the ratio between the circumference of a circle and its diameter, with the diameter equaling twice the radius
+ With C for circumference and 2*r* for diameter we have:
$$ \pi = \frac{C}{2r} \\ C = 2{\pi}{r} $$
+ From this it is easy to see that 2*&pi;* is the ratio between the circumference and the radius:
$$ 2{\pi} = \frac{C}{r} $$
+ We can also call the circumference the full *arc length* of a circle
+ What if we only want a section of the *arc length*?
+ This allows us to introduce the *radians*
+ Here we have *arc length*, *s*, as a section of the circumference and the angle made by two lines from the center to the circumference so as to enclose the section of *arc length*, the angle in radians

+ Note that in SI units, both arc length, *s*, and radius, *r*, are in meters and that their ratio, *radians*, is dimensionless

## Angular velocity

+ *Angular velocity* is the number of radians change per unit time
+ On average (or better still, for constant angular velocity) we have:
$$ \omega = \frac{\Delta \theta}{t} $$
+ Let's write *&Delta;&theta;* simply as *&theta;*; it is after all just an angle as the radius sweeps around (counterclockwise for positive), with the final angle the difference between the end and the beginning angle as taken from the zero radian line, which is defined as on the *x*-axis
+ For uniform circular motion and at a static point in time (*&omega;* and *&theta;* are constants) we can derive an equation:
$$ \omega =\frac { d\theta  }{ dt } \\ d\theta =\omega dt\\ \int { d\theta  } =\int { \omega  } dt\\ \theta =\omega t+\phi  $$
+ Here *&phi;* is the constant of integration
    + It is simply a starting angle

## Tangential velocity

+ In a specific time, *t*, a particle on the circumference will travel a certain arc length, *s*
+ From our standard equation for velocity being change in position over time we have:
$$ v=\frac{s}{t} \\ v=\frac{{r}{\theta}}{t} \\ v=r{\omega} $$
+ In vector form this will be:
$$ \underline { v } =\underline { \omega  } \times \underline { r }  $$

+ The period, *T*, is the time taken for a particle to complete the full circumference
+ The frequency is the number of *full circles* per second
+ We have:
$$ \omega =\frac { \theta  }{ t } =\frac { 2\pi  }{ T } \\ T=\frac { 2\pi  }{ \omega  } \\ \nu =\frac { 1 }{ T } \\ \omega =2\pi \nu  $$

## Turning circular motion into oscillatory motion and into periodic functions

+ In your mind's eye consider a point on the outer edge of a stationary turning wheel
+ It goes around and around and around and ...
+ Now turn the wheel so that you look at it edge on
+ That same point on the outer edge now goes periodically up and and down and up and ...
+ It is like a mass on a spring
+ In most introductory physics classes you may have seen a spray can attached between two springs that is made to oscillate up and down
+ As it sprays, a roll of paper is pulled in front of the spray can, producing a sine or cosine wave
+ This tells us that there is a relationship between circular motion, oscillatory motion and periodic functions

## The one-dimensional wave moving at constant velocity

+ We know that a wave can move, just go to the beach and have a look
+ The *medium* through which the wave moves does not move though
+ Think of a cork in the water, which just bobs up and down

+ We will set up our wave as a function of position and time:
$$ \Psi\left(x,t\right) $$
+ Furthermore, let's set it up akin to a cosine wave, with a maximum *&Psi;* value at *x*=0, when *t*=0; i.e. whatever is inside the parentheses must equal zero
+ Consider now the wave traveling at a constant velocity, *v*, and remember that for this, distance, *x* is velocity times time, *t*
+ It is the peak (same as at *x*=0 and *t*=0) that has moved and we must still have whatever is in the parentheses to be zero
+ We are now at a new *x* position, though, with *x*=*vt* and thus we must have:
$$ x-vt=0 $$
+ With a wave propagation to the right we have:
$$ \Psi\left(x,t\right)=\Psi\left(x-vt\right) $$
+ For the same arguments, we will have the wave traveling to the left to be:
$$ \Psi\left(x,t\right)=\Psi\left(x+vt\right) $$

### The one-dimensional time-independent wave equation

+ Let's play with some calculus and see how the wave function above changes with position, i.e. keeping time, *t*, constant
+ For this we need partial derivatives
+ First, to make it easier, let's introduce a new variable, *u*:
$$ u=x-vt\\ \therefore \quad \frac { \partial u }{ \partial x } =1 $$
+ With the chain rule we now have the following:
$$ \frac { \partial \Psi \left( x,t \right)  }{ \partial x } =\frac { \partial \Psi \left( u \right)  }{ \partial u }  $$

+ Calculus is so much fun, let's take the second derivative:
$$ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { x }^{ 2 } } =\frac { \partial  }{ \partial x } \left[ \frac { \partial \Psi \left( x,t \right)  }{ \partial x }  \right] =\frac { \partial  }{ \partial x } \left[ \frac { \partial \Psi \left( u \right)  }{ \partial u }  \right] \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { x }^{ 2 } } =\frac { { \partial  }^{ 2 }\Psi \left( u \right)  }{ \partial { u }^{ 2 } } \frac { \partial u }{ \partial x } \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { x }^{ 2 } } =\frac { { \partial  }^{ 2 }\Psi \left( u \right)  }{ \partial { u }^{ 2 } }  $$

### The distance-independent wave equation

+ Let's see what happens to the first and the second derivatives with respect to time, keeping *x* constant:
$$ \frac { \partial u }{ \partial t } =-v\\ \frac { \partial \Psi \left( x,t \right)  }{ \partial t } =\frac { \partial \Psi \left( u \right)  }{ \partial t } \frac { \partial u }{ \partial t } \\ \frac { \partial \Psi \left( x,t \right)  }{ \partial t } =-v\frac { \partial \Psi \left( u \right)  }{ \partial t } \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { t }^{ 2 } } =\frac { \partial  }{ \partial x } \left[ \frac { \partial \Psi \left( x,t \right)  }{ \partial t }  \right] =\frac { \partial  }{ \partial x } \left[ -v\frac { \partial \Psi \left( u \right)  }{ \partial t }  \right] \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { t }^{ 2 } } =-v\frac { { \partial  }^{ 2 }\Psi \left( u \right)  }{ \partial { u }^{ 2 } } \frac { \partial u }{ \partial t } \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { t }^{ 2 } } ={ v }^{ 2 }\frac { { \partial  }^{ 2 }\Psi \left( u \right)  }{ \partial { u }^{ 2 } }  $$

### The classical wave equation for a one-dimensional wave

+ We can combine our results above to get the following:
$$ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { t }^{ 2 } } ={ v }^{ 2 }\frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { x }^{ 2 } } \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { x }^{ 2 } } =\frac { 1 }{ { v }^{ 2 } } \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { t }^{ 2 } } \\ \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { x }^{ 2 } } -\frac { 1 }{ { v }^{ 2 } } \frac { { \partial  }^{ 2 }\Psi \left( x,t \right)  }{ \partial { t }^{ 2 } } =0 $$

+ This holds for movement both forward and backward (we are squaring), whereas we had the difficult situation of *x*-*vt* for forward (positive *x*) movement and *x*+*vt* for backward movement

## The sinusoidal or harmonic wave function

+ From all our work above, let's consider:
$$ \Psi\left(x,t\right)=A\cos{\left[k{\left(x-vt\right)}\right]} $$
+ The only new term being *k*, which we will call the *angular wavenumber*

+ Because we chose the cosine function (with a maximum when the argument is zero), we have *&Psi;* at a maximum here and equal to A, the amplitude
+ The difference in *&Psi;* from peak to through is 2A
+ Our wave function above also has a wavelength, *&lambda;*
+ The value of *&Psi;* will repeat every *&lambda;* multiple of our argument
    + For a normal cosine function we have:
$$ \cos { \left( \theta  \right)  } =\cos { \left( \theta +n2\pi  \right)  } ,\quad n\in \left( 0,1,2,3,\dots  \right)  $$
+ We can fit in more
    + In the graphs below we will have two functions:
    $$ \cos\left(\theta\right) \\ \cos\left(4\theta\right) $$

```python
from numpy import linspace, cos, pi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = linspace(0, 4 * pi, 100)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(x, cos(x), '-')
plt.show();
```

```python
plt.figure(figsize = (8, 6))
plt.plot(x, cos(4 * x), 'r-')
plt.show();
```

+ Above I used the constant value of four, but we can use any constant
+ In considering *&Psi;*, we repeat it's value for every &plusmn;*&lambda;*
    + So, when *x* goes from *x*=0 to *x* = &plusmn;*&lambda;* we have the same value for *&Psi;*
+ For the cosine function, this is akin to a constant multiple of 2*&pi;*
+ We call this constant the angular wavenumber, *k*
$$ k\lambda=2\pi \\ k=\frac{2\pi}{\lambda} $$
+ We can also explain it by considering *&Psi;* at *t*=0
$$ \Psi \left( x,0 \right) =A\cos { \left( kx \right)  } $$
+ We just noted that *&Psi;* remains the same if *x*=&plusmn;*&lambda;*, so we have:
$$ \Psi \left( x,0 \right) =A\cos { \left( k\lambda \right)  } $$
+ Back in the world of trigonometry this would move us on 2*&pi;* radians and again we have:
$$ k\lambda=2\pi \\ k=\frac{2\pi}{\lambda} $$

+ Please note that this is a simplification for one-dimension
+ In more than one dimension we have the propagation vector:
$$ \underline{k} \\ \underline{k} \cdot \underline{r} $$
+ The *wavenumber*  represents the number of wavelengths in a distance of one meter and is the reciprocal of the wavelength:
$$ \frac{1}{\lambda} $$

### Derivation of *&Psi;(x,t)*

+ Consider the following:
$$ \Psi \left( x,t \right) =A\cos { \left( kx-kvt \right)  } \\ \Psi \left( x,0 \right) =A\cos { \left( kx \right)  } \\ \Psi \left( 0,t \right) =A\cos { \left( -kvt \right)  } =A\cos { \left( kvt \right)  }  $$

+ The period, *T*, is the time taken between the successive passing of two adjacent points on the wave, i.e. peak-to-peak
+ In the trigonometric view, this would be the time taken to complete 2*&pi;* radians
+ Now we have:
$$ kvT=2\pi $$
+ Since frequency, *&nu;* is the reciprocal of the period, *T*, we have:
$$ \frac{kv}{\nu}=2\pi \\ \because {k}=\frac{2\pi}{\lambda} \\ \frac{2\pi}{\lambda} \frac{v}{\nu}=2\pi \\ \therefore v={\nu}{\lambda} $$

+ We can play with some algebra and finally we end up with the wave equation that is most familiar:
$$ kv=k{\nu}{\lambda} \\ kv=\frac{2\pi}{\lambda}{\nu}{\lambda} \\ kv=2\pi\nu=\omega \\ \because \Psi \left( x,t \right) =A\cos { \left( kx-kvt \right)  } \\ \Psi \left( x,t \right) =A\cos { \left( kx-\omega t \right)  }  $$
